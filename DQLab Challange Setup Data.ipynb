```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
advisedby = pd.read_csv('advisedBy.csv',sep = ";" )
taughtby = pd.read_csv('taughtBy.csv', sep = ";")
course = pd.read_csv('course.csv', sep = ";")
person = pd.read_csv('person.csv', sep = ";")
```

```python
df1 = person.set_index('p_id').join(taughtby.set_index('p_id'))
df1.head()
```

```python
df2 = df1.join(advisedby.set_index('p_id'))
df2.head()
```

```python
df2.reset_index(inplace=True)
```

```python
df3= df2.set_index('course_id').join(course.set_index('course_id'))
```
